## work 1

作业1：用4阶龙格-库塔法求解周期驱动下的单摆的运动

\begin{equation}
\begin{array}{l}
\frac{d^{2} \theta}{d t^{2}}+q \frac{d \theta}{d t}+\sin \theta=b \cos \omega_{0} t \\
y_{1}=\theta \text { and } y_{2}=\omega=d \theta / d t \\
\frac{d y_{1}}{d t}=y_{2} \\
\frac{d y_{2}}{d t}=-q y_{2}-\sin y_{1}+b \cos \omega_{0} t
\end{array}
\end{equation}

\begin{equation}
\begin{array}{l}
{y}_{i+1}={y}_{i}+\frac{1}{6}\left({c}_{1}+2 {c}_{2}+2 {c}_{3}+{c}_{4}\right) \\
{c}_{1}=\tau {g}\left({y}_{i}, t_{i}\right) \\
{c}_{2}=\tau {g}\left({y}_{i}+\frac{{c}_{1}}{2}, t_{i}+\frac{\tau}{2}\right) \\
{c}_{3}=\tau {g}\left({y}_{i}+\frac{{c}_{2}}{2}, t_{i}+\frac{\tau}{2}\right) \\
{c}_{4}=\tau {g}\left({y}_{i}+{c}_{3}, t_{i}+\tau\right)
\end{array}
\end{equation}

1. $ω_0=2/3，q=0.5, b=0.9, y_1(0)=0, y_2(0)=0$
2. $ω_0=2/3，q=0.5, b=1.15, y_1(0)=0, y_2(0)=0$

分别画出$Θ(t)$, t 从0 到 $20T_0$ 的演化轨迹，$T_0=2π/ω_0$
同时比较$dt=0.1$和$dt=0.01$这两种情况的结果

```python
def gxyt(y,t):
    f1 = y[1]
    f2 = -q*y[1] - np.sin(y[0]) + b*np.cos(omega*t)
    return np.array([f1,f2])

q = 0.5
omega = 2/3
peri_t = 2*np.pi/omega
def s_pen(tau):
    bushu = peri_t*20/tau
    t = [0,]
    y=np.array([0.01,0])
    yn = [y[0],]
    for i in range(int(bushu)):
        c11 = gxyt(y,t[-1])*tau
        c12 = gxyt(y + c11*0.5,t[-1] + tau*0.5)*tau
        c13 = gxyt(y + c12*0.5,t[-1] + tau*0.5)*tau
        c14 = gxyt(y + c13,t[-1] + tau)*tau
        dy1 = (c11+2*c12+2*c13+c14)/6.

        y = y + dy1
        t.append(t[-1] + tau)
        yn.append(y[0])
    return t,yn
```

```python
fig,axs=plt.subplots(2,1,figsize=(12,5),dpi=166)

bs=[0.9,1.15]
for i in range(2):
    ax=axs[i]
    b=bs[i]
    for dt in [0.01,0.1]:
        t,y1=s_pen(dt)
        # ax.plot(t,np.array(y1)%(2*np.pi),label=r'$b=%.2f$'%b)
        ax.plot(t,np.array(y1),label=r'$b=%.2f,dt=%.2f$'%(b,dt),alpha=0.6)
    ax.set_xlabel('time t')
    ax.set_ylabel(r'$\theta$')
    ax.legend(loc = 'upper left')
    labels=[0,r'$5T_0$',r'$10T_0$',r'$15T_0$',r'$20T_0$']
    ax.set_xticks([0,5*peri_t,10*peri_t,15*peri_t,20*peri_t])
    ax.set_xticklabels(labels)

fig.tight_layout()
plt.show()
```

## work 2

作业2：范德波尔振荡器（Van der Pol oscillator）是荷兰物理学家巴尔
塔萨·范·德·波尔在1927年发现的真空管放大器的极限环振荡现象[1]。极限
环振荡可以用下列非线性微分方程表示

\begin{equation}
\frac{d^{2} x}{d t^{2}}-\alpha\left(1-x^{2}\right) \frac{d x}{d t}+x=0
\end{equation}

1. $α=3, dt=0.01,x(0)=0, v(0)=0$
1. $α=3, dt=0.01,x(0)=2, v(0)=0$

分别画出x(t), t 从0 到 50的演化轨迹

\begin{equation}
\dot x = v \\
\dot v = \alpha (1-x^2)v-x \\
\end{equation}

```python
def slope_v(x,t):
    xd = x[1]
    a=3
    vd = a *(1-x[0]**2)*x[1] - x[0]
    return np.array([xd, vd])

def vdp(x0):
    dt=0.01
    x=np.array([x0,0])
    ts=[0,]

    xn=[x[0],]
    bushu = int(50/dt)
    t = 0
    for i in range(bushu):
        k1 = slope_v(x,t)*dt
        k2 = slope_v(x + k1*0.5, t + dt*0.5)*dt
        k3 = slope_v(x + k2*0.5, t + dt*0.5)*dt
        k4 = slope_v(x + k3, t + dt)*dt
        dx = (k1+k2*2+k3*2+k4)/6

        x = x + dx
        t += dt
        ts.append(t)
        xn.append(x[0])
    return ts,xn
```

```python
fig,axs=plt.subplots(4,1,figsize=(10,6),dpi=166)

x0s=[0,0.1,0.5,2]
for i in range(4):
    ax=axs[i]
    x0=x0s[i]
    ts,xn = vdp(x0)
    ax.plot(ts,xn,label=r'$x_0=%.2f$'%x0)

    ax.set_xlabel('time t')
    ax.legend(loc = 'upper right')

fig.tight_layout()
plt.show()
```

作业3：数值求解一维谐振子的定态薛定谔方程

\begin{equation}
-\frac{\hbar^{2}}{2 m} \frac{d^{2} \phi(x)}{d x^{2}}+V(x) \phi(x)=\varepsilon \phi(x)
\end{equation}

取$m=1, V(x)=0.5x^2$, 在-4<x<4的范围内，画出基态波函数随空间的分布，同时求出基态能量

\begin{equation}
\frac{d\phi}{dx} = \phi ' \\
\frac{d\phi '}{dx} = -\frac{2m}{\hbar^2}(\varepsilon - V(x))\phi \\
\end{equation}

```python
def slope_sch(phi,x,e):
    moh2 = 1
    phip = phi[1]
    phipp = (0.5*x**2-e)*phi[0]*2*moh2
    return np.array([phip, phipp])

def sch(e):
    dx=0.01
    wf=np.array([0,0.01])
    xs=[-4,]

    wfn=[wf[0],]
    bushu = int(8/dx)
    x = -4
    for i in range(bushu):
        k1 = slope_sch(wf,x,e)*dx
        k2 = slope_sch(wf + k1*0.5, x + dx*0.5,e)*dx
        k3 = slope_sch(wf + k2*0.5, x + dx*0.5,e)*dx
        k4 = slope_sch(wf + k3, x + dx,e)*dx
        dwf = (k1+k2*2+k3*2+k4)/6

        wf = wf + dwf
        x += dx
        xs.append(x)
        wfn.append(wf[0])
    return xs,wfn

def wavefunc(e):
    xs,wfn = sch(e)
    phi4=wfn[-1]
    return phi4

def fe(teste):
    phi4 = 0
    fe = wavefunc(teste) - phi4
    return fe

def eigene(start=0):
    en = start
    en1 = 0.1 + start
    while ((en1-en)>1e-5):
        en2 = en1 - (en1 - en)*fe(en1)/(fe(en1)-fe(en))
        en,en1 = en1,en2
    return en1
```

```python
fig,ax=plt.subplots(figsize=(10,3),dpi=166)

for i in range(3):
    e = eigene(i)
    xs,wfn = sch(e)
    temp=np.array(wfn)
    nom=(temp**2*0.01).sum()
    wfn = temp /np.sqrt(nom)
    ax.plot(xs,wfn,label=r'$\varepsilon=%.2f$'%e)

ax.set_xlabel('x')
ax.set_ylabel(r'$\phi(x)$')
ax.set_title('wave function ')
ax.legend(loc = 'upper right')

fig.tight_layout()
plt.show()
```

## work 4

作业4：用梯度法（最速下降法）求解非线性方程组的一组根：

\begin{equation}
\begin{array}{l}
x-5 y^{2}+7 z^{2}+12=0 \\
3 x y+x z-11 x=0 \\
2 y z+40 x=0
\end{array}
\end{equation}


```python
def func(x,y,z):
    f1=x-5*y**2+7*z**2+12
    f2=3*x*y+x*z-11*x
    f3=2*y*z+40*x
    f=f1**2+f2**2+f3**2
    return f

def funcx(x,y,z,h):
    funcx=(func(x+h,y,z)-func(x-h,y,z))/(2*h)
    return funcx

def funcy(x,y,z,h):
    funcy=(func(x,y+h,z)-func(x,y-h,z))/(2*h)
    return funcy

def funcz(x,y,z,h):
    funcz=(func(x,y,z+h)-func(x,y,z-h))/(2*h)
    return funcz
```

```python
xi=np.array([0,0,0])
h=0.0001

grad=np.array([1,1,1])
while (h*(grad**2).mean() > 1e-10):
    x,y,z=xi[:]
    grad = np.array([funcx(x,y,z,h/100),funcy(x,y,z,h/100),funcz(x,y,z,h/100)])
    xi = xi - h*grad
print('x,y,z=',xi)
```
