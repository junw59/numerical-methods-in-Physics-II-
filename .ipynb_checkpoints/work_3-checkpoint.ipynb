## work 1

作业3：数值求解一维谐振子的定态薛定谔方程

\begin{equation}
-\frac{\hbar^{2}}{2 m} \frac{d^{2} \phi(x)}{d x^{2}}+V(x) \phi(x)=\varepsilon \phi(x)
\end{equation}

取$m=1, V(x)=0.5x^2$, 在-4<x<4的范围内，画出基态波函数随空间的分布，同时求出基态能量

\begin{equation}
\frac{d\phi}{dx} = \phi ' \\
\frac{d\phi '}{dx} = -\frac{2m}{\hbar^2}(\varepsilon - V(x))\phi \\
\end{equation}

```python
def slope_sch(phi,x,e):
    moh2 = 1
    phip = phi[1]
    phipp = (0.5*x**2-e)*phi[0]*2*moh2
    return np.array([phip, phipp])

def sch(e):
    dx=0.01
    wf=np.array([0,0.01])
    xs=[-4,]

    wfn=[wf[0],]
    bushu = int(8/dx)
    x = -4
    for i in range(bushu):
        k1 = slope_sch(wf,x,e)*dx
        k2 = slope_sch(wf + k1*0.5, x + dx*0.5,e)*dx
        k3 = slope_sch(wf + k2*0.5, x + dx*0.5,e)*dx
        k4 = slope_sch(wf + k3, x + dx,e)*dx
        dwf = (k1+k2*2+k3*2+k4)/6

        wf = wf + dwf
        x += dx
        xs.append(x)
        wfn.append(wf[0])
    return xs,wfn

def fe(teste):
    xs,wfn = sch(teste)
    fe = wfn[-1] - wfn[0]
    return fe

def eigene(start=0):
    en = start
    en1 = 0.1 + start
    while ((en1-en)>1e-6):
        en2 = en1 - (en1 - en)*fe(en1)/(fe(en1)-fe(en))
        en,en1 = en1,en2
    xs,wfn = sch(en1)
    temp=np.array(wfn)
    nom=(temp**2*0.01).sum()
    wfn = temp /np.sqrt(nom)
    return en1,xs,wfn
```

```python
fig,ax=plt.subplots(figsize=(10,3),dpi=166)

for i in range(3):
    e,xs,wfn = eigene(i)
    ax.plot(xs,wfn,label=r'$\varepsilon=%.2f$'%e)

ax.set_xlabel('x')
ax.set_ylabel(r'$\phi(x)$')
ax.set_title('wave function ')
ax.legend(loc = 'upper right')

fig.tight_layout()
plt.show()
```
